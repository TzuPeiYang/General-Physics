```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def forward_difference(f, x, h=1e-4):
    return (f(x + h) - f(x))/h

def central_difference(f, x, h=1e-4):
    return (f(x + h/2) - f(x - h/2))/h
```

```python
def f(x):
    return x**3.
```

```python
print(central_difference(f, 5))
print(forward_difference(f, 5))
```

```python
error_list = []
for i in range(1, 15):
    error_list.append(np.log10(abs(75 - central_difference(f, 5, h=10**-i))))
    
fig = plt.figure(figsize=(8, 4), dpi=200)
plt.plot(range(-1, -15, -1), error_list, 'o')
plt.xlim(-15, 0)
plt.xlabel("$log(h)$")
plt.ylabel("$log(error)$")
plt.xticks(range(-15, 0, 2))
fig.savefig("123.png")
plt.show()
```
