```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def euler_method(f, r_init, time_interval, h): #(方程式，初始向量，時間區間，時間變化量)
    sim_time = time_interval[1] - time_interval[0]
    iters = int(sim_time / h)
    
    r_list = [np.copy(r_init)]
    t_list = [time_interval[0]]
    
    r_new = r_init
    t = time_interval[0]
    for i in range(iters):
        r_new += f(r_new, t) * h
        t += h
        
        r_list.append(np.copy(r_new))
        t_list.append(t)
    
    return np.array(r_list), np.array(t_list)
```

```python
def x(r, t): # 一維拋體
    g = 9.8
    return np.array([r[1], -g]) # r = [x, v], 微分後 x = [v, a]
```

```python
r_init = np.array([100, 0], dtype=float)
time_interval = [0, 4]
h = 1e-1

r, t = euler_method(x, r_init, time_interval, h)

plt.plot(t, r[:, 0])
plt.xlabel('time(s)')
plt.ylabel('height(m)')
plt.xlim(0, 4)
plt.show()
```

```python
def small_angle_approx(t, A=50):
    l = 1
    g = 9.8
    return A/180*np.pi*np.cos(np.sqrt(g/l)*t)
```

```python
def f(r, t):
    l = 1
    g = 9.8
    return np.array([r[1], -g/l*np.sin(r[0])])
```

```python
r_init = np.array([50/180*np.pi, 0], dtype=float)
time_interval = [0, 10]
h = 1e-4

r, t = euler_method(f, r_init, time_interval, h)

plt.plot(t, r[:, 0])
plt.plot(t, small_angle_approx(t))
plt.xlabel('time(s)')
plt.ylabel(r'$\theta$')
plt.xlim(0, 10)
plt.show()
```

```python
r_init = np.array([10/180*np.pi, 0], dtype=float)
time_interval = [0, 10]
h = 1e-4

r, t = euler_method(f, r_init, time_interval, h)

plt.plot(t, r[:, 0])
plt.plot(t, small_angle_approx(t, A=10))
plt.xlabel('time(s)')
plt.ylabel(r'$\theta$')
plt.xlim(0, 10)
plt.show()
```
