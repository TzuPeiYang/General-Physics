```python
def factorial(n):
    product = 1
    for i in range(1, n+1):
        product *= i
    return product
```

```python
print(factorial(5))
print(factorial(0))
```

```python
def factorial(n):
    if n <= 1:
        return 1
    return n*factorial(n - 1)
```

```python
print(factorial(5))
```
